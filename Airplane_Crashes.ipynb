```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importando o CSV e printando em tela o DataFrame
df = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")

display(df.head())
```

```python
#Agrupando por quantidade de passageiros abordo do avi√£o
abordo = df.groupby('Aboard')['Aboard'].count()
abordo.dropna()
print(abordo.index)
```

```python
print(abordo.values)
```

```python
#Plotando gr√°fico que mostra 
x = abordo.index
y = abordo.values

dado = [10**i for i in range(4)]

fig, ax = plt.subplots(ncols = 2, nrows = 1, figsize = (16,8))
ax[0].scatter(x,y)
ax[0].plot(x,y)
ax[1].bar(x,y)

ax[0].set(xlim=(0, 8), xticks=np.arange(1, 16))
ax[1].set(xlim=(0, 8), xticks=np.arange(1, 16))

ax[0].set_title('Quedas X Quantidade de Passageiros Abordo')

for i in range(0, len(abordo)):
    plt.annotate(y[i], (i, y[i]), ha='center')
```
